```python
import pandas as pd
import numpy as np
import yfinance as yf
from datetime import timedelta

import seaborn as sn
```

```python
ticker_list = ['SPY','QQQ', 'EEM', 'UUP','GLD', 'TLT', 'IYR', 'EFA', 'IWM', 'USO']
lookback = 126
```

```python
prices = pd.DataFrame(columns=ticker_list)
```

## Obtain Prices <a id='prices'>

```python
for ticker in ticker_list:
    prices[ticker] = yf.download(ticker, '2015-01-01', '2018-12-31')['Adj Close']

prices.sort_values(by=['Date'], inplace=True, ascending=False)
prices.head()
```

## Create Return Arrays <a id='returns'>

```python
# Find a return
np.log(prices['SPY'][0]/prices['SPY'][1])
```

```python
# Create a return array
price = prices['SPY']
np.log(price / price.shift(-1))
```

```python
def stock_return(prices, daily_periods):
    return np.log(prices / prices.shift(-daily_periods))
```

```python
returns = pd.DataFrame(columns=ticker_list)

for ticker in ticker_list:
    returns[ticker] = stock_return(prices[ticker], 1)

returns = returns[:lookback]
returns
```

```python
for ticker in ticker_list:
    returns[ticker + '-1'] = returns[ticker] * -1

returns
```

```python
## Filter to positive returns

direction = []

for ticker in ticker_list:
    if prices[ticker][0] > prices[ticker][lookback]:
        direction.append(True)
    else:
        direction.append(False)

## Adjust Return m
direction = np.array(direction)
direction = np.concatenate((direction, np.logical_not(direction)))
direction
```

```python
## Adjusted Returns Table
returns = returns[returns.columns[direction]]
```

## Create Correlation Matrix

```python
correlations = returns.corr()
sn.heatmap(correlations, annot=True)
```

## Initialize Portfolio Weights

```python
weights = np.random.rand(1, len(returns.columns))
weights
```

## Create Weights Matrix

```python
weight_matrix = weights.T.dot(weights)
sn.heatmap(weight_matrix, annot=True)
```

## Find Intra-Portfolio Correlation

```python
ipc_matrix = weight_matrix * correlations
ipc = ipc_matrix.to_numpy().sum()
ipc
```

## Minimize IPC

```python
import scipy.optimize as sco
```

```python
def IPC(weights, correlation_matrix):
    #print(weights)
    weight_matrix = np.dot(weights.T,weights)
    #print(weight_matrix)
    ipc_matrix = weight_matrix * correlation_matrix
    #print(ipc_matrix)
    print(ipc_matrix.to_numpy().sum())
    return ipc_matrix.to_numpy().sum()
```

```python
initial_weights = np.random.rand(1,len(returns.columns))
def min_ipc(correlation_matrix):
    num_assets = len(ticker_list)
    args = (correlation_matrix)
    constraints = ({'type': 'eq', 'fun': lambda x: np.sum(x) - 1})
    bound = (0.0,1.0)
    bounds = tuple(bound for asset in range(num_assets))
    
    result = sco.minimize(IPC, x0=initial_weights, args=args,
                        method='SLSQP', bounds=bounds, constraints=constraints)
    return result
```

```python
optimized = min_ipc(correlations)
optimized
print(IPC(optimized.x,correlations))
print(optimized.x)
```
